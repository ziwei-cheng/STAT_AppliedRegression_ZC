#### (a) Suppose that you'd like to conduct a hypothesis test at the $\alpha = 0.02$ level to decide whether the true mean temperature (C) at harvest is less than 18 degrees C. State the null and alternative hypotheses   (10 PTS) .

Answer:

The null hypothesis is that the true mean temperature (C) at harvest is 18 degrees C. The alternative hypothesis is that the true mean temperature at harvest is less than 18 degrees C. 

#### (b) Conduct the appropriate hypothesis test using the p-value method. Clearly state your conclusion. (10 PTS)

```r
alpha = 0.02; cv = qnorm(1-alpha/2)
xbar = mean(wine$h.temp); 
s = sd(wine$h.temp);
n = length(wine$h.temp)
z = (xbar - 18)/(s/sqrt(n)); z
pvalue = pnorm(z); 
cat("The p-value is", pvalue, ".")
```

Answer:

The p-value is 0.0015 and smaller than 0.02. It means that we can reject the null hypothesis. It means we reject the hypothesis that the  true mean temperature at harvest is 18 degrees C.
