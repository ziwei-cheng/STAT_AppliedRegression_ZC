## Problem #1

(Same setup as theoretical problem 3): The random variable $X$ can take on the values 0,1,2, and 3 with probabilities $P(X = 0) = p^3$, $P(X = 1) = (1-p)p^2$, $P(X = 2) = (1-p)^2$, and $P(X = 3) = 2p(1-p)$, where $0 < p < 1$. Graph the log likelihood function in R and place a  vertical line at the MLE found in theoretical problem 3. Does it look like the correct maximizer?


```r
l = function(p){
    l = 24*log(p^3)+54*log(p^2-p^3) + 32*log((1-p)^2) + 40*log(2*p-2*p^2)
    return(l)
}

p = seq(0,1,length = 500);
plot(p,l(p), type = "l")
```
