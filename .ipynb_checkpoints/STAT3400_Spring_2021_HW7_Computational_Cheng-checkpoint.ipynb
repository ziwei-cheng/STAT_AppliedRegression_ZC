### Problem #1

#### (a) Using the divusa data, fit a regression of with divorce as the response and unemployed, femlab, marriage, birth, and military as predictors. Compute the condition number and interpret its meaning.


```r
library(faraway)
library(leaps)
```

```r
data(divusa)
```

```r
lmod=lm(divorce~unemployed+femlab+marriage+birth+military,data=divusa)
summary(lmod)
```

```r
#kappa(lmod)
 x <- model.matrix(lmod)
 e <- eigen(t(x) %*% x)
 low = min(e$values)
 high = max(e$values)
 round(sqrt(high/low), 2)
```

Answer:The condition number is larger than 30 and it suggests colinearity.

#### (b) For the same model, compute the VIFs. Is there evidence that collinearity causes some predictors not to be significant? Explain your answer.

```r
vif(lmod)
```

Answer:
VIF greater than 10 are evidence of collinearity. No evidence of colinearity in the predictors

#### (c) Does the removal of insignificant predictors from the model reduce the collinearity? Investigate.

```r
lmod1=lm(divorce~femlab+marriage+birth,data=divusa)
summary(lmod1)
```

```r
sum(vif(lmod1))
vif(lmod1)
```

```r
 x <- model.matrix(lmod1)
 e <- eigen(t(x) %*% x)
 low = min(e$values)
 high = max(e$values)
 round(sqrt(high/low), 2)
```

Answer:

The removal of insignificant predictors from the model reduce the collinearity because vifs decreased and condition number also decreased. 

### Problem #2

Use the prostate data with lpsa as the response and the other variables as predictors. Implement the following variable selection methods to determine the "best" model:

1. Backward Elimination
2. AIC
3. Adjusted $R^2$.


```r
#backward elimination, this method fits lpsa~lcavol+lweight+svi (3 predictors)
lmod2=lm(lpsa~., data=prostate)
summary(lmod2)
```

```r
lmod_back=(lm(lpsa~lcavol+lweight+svi, data=prostate))
summary(lmod_back)
```

```r
#AIC, the model fits lpsa~lcavol+lweight+age+lbph+svi (5 predictors)
n = length(prostate$lpsa)
b = regsubsets(lpsa ~ ., data = prostate)
rs = summary(b); aic = n*log(rs$rss/n) + 2*(2:9); i = which.min(aic)
plot(aic ~ I(1:8), ylab = "AIC", xlab = "Number of Predictors", 
     type = "o")

aic = prostate[,rs$which[i,-1]];
lmodAIC = lm(lpsa ~ ., data = aic); summary(lmodAIC)
```

```r
#R^2, the model fits lpsa~lcavol+lweight+age+lbph+svi+lcp+pgg45 (7 predictors)
plot(I(1:8), rs$adjr2, type = "o", ylab = expression(Ra^2), xlab = "Number of Predictors", )
summary(b)
which.max(rs$adjr2)
lmodr2 = lm(lpsa ~ .-gleason, data = prostate)
summary(lmodr2)
```

```r
par(mfrow = c(2,2))
plot(lmod2,1:2)
plot(lmodAIC,1:2)
plot(lmodr2,1:2)
```

Answer:

None of the models violated constant variance, linearity, or normality. Also, all three models have similar R^2 values (62%), so I think the simplest model (lpsa~lcavol+lweight+svi (3 predictors)) is the best model.

### Problem #3

Use the teengamb data with gamble as the response and the other variables as predictors. Implement the following variable selection methods to determine the "best" model:

1. Backward Elimination
2. AIC
3. Adjusted $R^2$.

```r
#backward elimination, fits gamble~sex+income (2 predictors)
lmod3=lm(gamble~., data=teengamb)
summary(lmod3)
```

```r
lmod3_back=lm(gamble~sex+income, data=teengamb)
summary(lmod3_back)
```

```r
#AIC,  fits gamble~sex+income+verbal (3 predictors)
n1 = length(teengamb$gamble)
b1 = regsubsets(gamble ~ ., data = teengamb)
rs1 = summary(b1); aic = n1*log(rs1$rss/n1) + 2*(2:5); i = which.min(aic)
plot(aic ~ I(1:4), ylab = "AIC", xlab = "Number of Predictors", 
     type = "o")

aic = teengamb[,rs1$which[i,-1]];

lmod3AIC = lm(gamble ~ ., data = aic); summary(lmod3AIC)
```

```r
#R^2, fits gamble~sex+income+verbal (3 predictors)
plot(1:4, rs1$adjr2, type = "o", ylab = expression(Ra^2), xlab = "Number of Predictors", )
summary(b1)
which.max(rs1$adjr2)

lmod3r2 = lm(gamble ~ .-status, data = teengamb)
summary(lmod3r2)
```

```r
par(mfrow = c(2,2))
plot(lmod3,1:2)
plot(lmod3AIC,1:2)
plot(lmod3r2,1:2)
```

Answer:

All three models have similar plots. Also, the last two models fits the same predictors and have a larger R^2, so I think the model (gamble~sex+income+verbal) is the best model.

### Problem #4

Use the divusa data with divorce as the response and the other variables as predictors. Implement the following variable selection methods to determine the "best" model:

1. Backward Elimination
2. AIC
3. Adjusted $R^2$.

```r
#backward elimination, fits divorce~year+femlab+marriage+birth+military (5 predictors)
lmod4=lm(divorce~., data=divusa)
summary(lmod4)
```

```r
lmod4_back=update(lmod4,.~.-unemployed)
summary(lmod4_back)
```

```r
#AIC, fits divorce~year+femlab+marriage+birth+military (5 predictors)
n2 = length(divusa$divorce)
b2 = regsubsets(divorce ~ ., data = divusa)
rs2 = summary(b2); aic = n2*log(rs2$rss/n2) + 2*(2:7); i = which.min(aic)
plot(aic ~ I(1:6), ylab = "AIC", xlab = "Number of Predictors", 
     type = "o")

aic = divusa[,rs2$which[i,2]];

rs2
```

```r
#R^2, fits divorce~year+femlab+marriage+birth+military (5 predictors)
plot(1:6, rs2$adjr2, type = "o", ylab = expression(Ra^2), xlab = "Number of Predictors", )
summary(b2)
which.max(rs2$adjr2)
```

Answer:

All three models fits the same predictors, (fits divorce~year+femlab+marriage+birth+military), so this is the best model.
