```r
calc_ci = function(data, percent){
  n = length(data)
  p_hat = mean(data)
  se = sqrt(p_hat * (1 - p_hat) / n)
  z_star = qnorm(p = (1-percent) / 2, lower.tail = FALSE)
  c(lower = p_hat - z_star * se, upper = p_hat + z_star * se)
}
```

```r
sum_0=0 
sum_1=0

#to simulate start value
p=8/23
#simulate one step forward probabilities
p00=1/4
p10=2/5
for(i in 1:10000){
    x<-runif(1,0,1)
    #determine which state to start
    if(x<p){
            #start at state 0
            x1<-runif(1,0,1)
            #one step forward
            if(x1<p00){
                sum_0=sum_0+1
            }
            else{
                sum_1=sum_1+1
            }
                
    }
    else{
            #start at state 1
            x2<-runif(1,0,1)
            #one step forward
            if(x2<p10){
                sum_0=sum_0+1
            }
            else{
                sum_1=sum_1+1
        }
                
    }
} 
sum_0/10000
sum_1/10000
```
